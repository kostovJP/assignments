```python
import matplotlib.pyplot as plt
import networkx as nx
G = nx.Graph()
for i in range(1,6):
    G.add_node(i)#adds nodes to the graph
G.nodes()#displays the list of all the nodes in the graph
G.add_edge(1,2)#adds an edge between the specified two nodes in the graph
G.add_edge(1,4)
G.add_edge(2,3)
G.add_edge(3,4)
G.add_edge(3,5)
G.add_edge(4,5)
G.edges()#displays a list of all the edges in the graph
nx.draw(G, with_labels = 1)#draws the graph and then we can display is using plt.show()
plt.show()
```

```python
Z = nx.complete_graph(10)#--->creates a complete graph having 10 vertices
Z.nodes()
Z.edges()

Z.order()#-->returns the number of vertices
Z.size()#--->returns the number of edges

nx.draw(Z, with_labels = 1)
plt.show()
```
