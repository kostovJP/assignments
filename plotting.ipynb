```python
plt.plot([1,2,3,4,5],[1,4,9,16,25], 'go') #g-->green, o-->points
plt.show()
plt.plot([1,2,3,4,5],[1,4,9,16,25], 'go-') #g-->green, o-->points
plt.show()
plt.plot([1,2,3,4,5],[1,4,9,16,25], 'y^-') #g-->green, o-->points, - --> lines, ^ --> triangles
plt.show()
plt.plot([1,2,3,4,5],[1,4,9,16,25], 'rs-') #g-->green, o-->points,s --> squares
plt.show()

#there are many more styles for representing the graphs
```

```python
plt.title("plot of numbers v/s their squares")
# plt.ylabel("squares of numbers")
# plt.xlabel("the numbers")
# plt.show()
plt.xlabel("the numbers", fontsize = 18, color = 'red')
plt.ylabel("sum of numbers", fontsize = 18, color = 'green')
# plt.show()
#the plt.axis([X1,X2,Y1,Y2]) takes a list of 4 values where X1-->starting of x axis, X2-->ending of X axis, Y1-->starting of Y axis, Y2-->starting of Y axis
plt.axis([0,6,0,30])
plt.plot([1,2,3,4,5],[1,4,9,16,25], 'rs-')
plt.show()
plt.plot([1,2,3,4,5],[1,4,9,16,25], 'rs-', linewidth = 3.0)#linewidth-->used to make the line thicker or thinner
plt.show()
```

```python
import numpy as np
#displaying multiple lines in one graph:
num = int(input("enter how many numbers you want: "))
x = [i for i in range(num)]
x = np.array(x)# converting the list to an array
print(x)
#since x is a list so opearations like ** are not allowed but to make it possible we must import numpy...
plt.plot(x, x**2, 'r-', x, x**3, 'b-', x, x**4, 'g-')
plt.xlabel('x-axis', color = 'red')
plt.ylabel('y-axis', color = 'red')
plt.show()
```
